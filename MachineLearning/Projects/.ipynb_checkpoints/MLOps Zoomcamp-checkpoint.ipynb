```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_data_histograms(df, figsize=(12, 10), ncols=3, nrows=None):
    """
    Creates a visually appealing histogram plot for each column in a pandas DataFrame.

    Args:
        df (pandas.DataFrame): The DataFrame containing the data to be visualized.
        figsize (tuple, optional): The desired figure size. Defaults to (12, 10).
        ncols (int, optional): The number of columns in the plot layout. Defaults to 3.
        nrows (int, optional): The number of rows in the plot layout.
                               If not provided, will be calculated based on the number of columns.

    Returns:
        matplotlib.figure.Figure: The generated figure object.
        numpy.ndarray: A 2D array of subplot axes for further customization.
    """

    nrows = nrows or (df.shape[1] // ncols + 1)  # Calculate nrows if not provided
    fig, axis = plt.subplots(nrows, ncols, figsize=figsize)

    # Iterate over columns and create histograms with minor adjustments
    for i, (col_name, col) in enumerate(df.items()):
        row, col_index = divmod(i, ncols)  # Efficiently calculate row and column index
        axis[row, col_index].hist(col)
        axis[row, col_index].set_title(col_name)  # Add column names as titles

    # Adjust layout and appearance for better readability (consider adding x/y labels, legends)
    fig.suptitle(f"Histograms of Columns in DataFrame", fontsize=14)
    fig.tight_layout()  # Adjust spacing between subplots

    return fig, axis

# Example usage
fig, axis = plot_data_histograms(df)
plt.show()
```

```python
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w = XTX_inv.dot(X.T).dot(y)
    
    return w[0], w[1:]

def prepare_X(df, fillna_value):
    df = df.fillna(fillna_value)
    X = df.values
    return X

def rmse(y, y_pred):
    error = y_pred - y
    mse = (error ** 2).mean()
    return np.sqrt(mse)
```
